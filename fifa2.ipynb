```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading data with pandas
fifa= pd.read_csv("players_20.csv")
```

```python
fifa.head(5)
```

```python
# Name of all columns
for cl in fifa.columns:
    print(cl)
```

```python
fifa.describe()
```

```python
# Number of rows and columns 
fifa.shape
```

```python
# Number of players for each country
fifa['nationality'].value_counts()
```

```python
# Top 8 countries with highest number of players
fifa["nationality"].value_counts()[0:8].keys()
```

```python
plt.figure(figsize=(12,6))
plt.bar(list(fifa["nationality"].value_counts()[0:8].keys()),list(fifa["nationality"].value_counts()[0:8]),color="orange");
plt.ylabel('Number of players')
plt.xlabel('Country')
plt.show
```

```python
# Highest paid players in the world
fifa2 = fifa[["short_name", "wage_eur"]]
fifa2.head()
```

```python
highest_salary = fifa2.sort_values(by="wage_eur", ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
plt.bar(highest_salary["short_name"][0:10], highest_salary["wage_eur"][0:10], color="#240076")
plt.xlabel("player")
plt.xticks(rotation='20')
plt.ylabel("Salary")
```

```python
# Brazilian players
fifa["nationality"]=="Brazil"
brazil = fifa[fifa["nationality"]=="Brazil"]
brazil.head(10)
```

```python
# Brazilian players with the highest shooting power
brazil_shooting = brazil.sort_values(by=['shooting'], ascending=False).head(10)
brazil_shooting
```

```python
# Top 5 highest shooting power in the world
player_shooting = fifa[["short_name", "shooting"]]
player_shooting.sort_values(by=["shooting"], ascending=False).head(5)
```

```python
# top 10 defenders in the world
player_defending = fifa[["short_name", 'defending', 'nationality', "club"]]
player_defending.sort_values(by='defending', ascending=False).head(5)
```

```python
# highest paid playerS in Real Madrid club
real_madrid = fifa[fifa["club"]=='Real Madrid']
real_madrid.sort_values(by=["wage_eur"], ascending=False).head(5)
```

```python
# Highest shooting power in Real Madrid
real_madrid.sort_values(by=["shooting"], ascending=False).head()
```
